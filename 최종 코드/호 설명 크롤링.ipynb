```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select 
import time

SCROLL_PAUSE_TIME = 3

# 웹 드라이버 동적 다운로드 방식
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())

# 관세청법령정보포탈 url 들어가기
driver.get('https://unipass.customs.go.kr/clip/index.do')
time.sleep(SCROLL_PAUSE_TIME)
# 세계HS 클릭
driver.find_element(By.LINK_TEXT, '세계HS').click()
time.sleep(SCROLL_PAUSE_TIME)

driver.find_element(By.LINK_TEXT, '속견표').click()
time.sleep(SCROLL_PAUSE_TIME)

driver.find_element(By.CSS_SELECTOR, '#tblLstBody > tr:nth-child(9) > td:nth-child(7) > a').click()
time.sleep(SCROLL_PAUSE_TIME)
# 84
#tblLstBody > tr:nth-child(9) > td:nth-child(6) > a
# 85
#tblLstBody > tr:nth-child(9) > td:nth-child(7) > a

driver.find_element(By.CSS_SELECTOR, '#tblLstBody > tr:nth-child(1) > td:nth-child(1) > a').click()
time.sleep(SCROLL_PAUSE_TIME)

hs_code = []
info_list = []
count = 0
for c in range(48):
    cnt = driver.find_elements(By.CSS_SELECTOR, '#tblLstBody > tr')
    print(len(cnt))
    for i in range(1, len(cnt)+1):
        time.sleep(SCROLL_PAUSE_TIME)
        code = driver.find_element(By.CSS_SELECTOR, '#tblLstBody > tr:nth-child(1) > td.textCenter.divClss_1.hlzone4 > a').text
        hs_code.append(code)
    
        info = driver.find_element(By.CSS_SELECTOR,f'#tblLstBody > tr:nth-child({i}) > td:nth-child(5) > a').text
        info_list.append(info)

        count += 1
        print(count, end=' ')
        
    driver.find_element(By.CSS_SELECTOR, '#searchVo > div.rightArea70 > div > div.paging.law_ryu > ul:nth-child(3) > li').click()
    time.sleep(SCROLL_PAUSE_TIME)

print('The end')
driver.close()
```

```python
import pandas as pd

data = {'hs code': hs_code, 'heading_info': info_list}
df = pd.DataFrame(data)
```
